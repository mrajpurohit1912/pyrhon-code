```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import seaborn as sns
plt.style.use('seaborn')
import datetime 
from pandas.plotting import register_matplotlib_converters
from statsmodels.graphics.tsaplots import plot_acf , plot_pacf
register_matplotlib_converters()
```

```python
df = pd.read_csv('ice_cream.csv',parse_dates = True,index_col = 0)
```

```python
df.rename(columns = {'IPN31152N':'price'},inplace = True)
```

```python
start = pd.to_datetime('2010-01-01')
```

```python
df = df[start:] 
```

```python
plt.figure(figsize = (12,8))
plt.plot(df)
plt.title("Chart pf adjf[")
plt.ylabel("nsdfasdl")
plt.legend(loc = 'best')

for year in range(2010,2020):
    plt.axvline(pd.to_datetime(str(year)+"-12-31"),linestyle = '--',alpha = 0.4)
```

```python
acf = plot_acf(df,lags = 100)
```

```python
acf = plot_pacf(df)
```

```python
yf = yf.download(tickers="MG",start = "2010-1-1",end = "2022-1-1")
```

```python
yf.drop(columns = {"Open","High","Low","Close","Volume"},inplace = True)
yf.rename(columns = {'Adj Close':'close'},inplace = True)
```

```python
yf.plot(figsize = (12,8))
plt.legend(loc = "upper right")

for year in range(2010,2022):
    plt.axvline(pd.to_datetime(str(year)+"-12-31"),linestyle = "--",alpha = 0.4)
    
plt.axhline(yf['close'].mean(),linestyle = '--',alpha = 0.4,color = 'maroon')
```
